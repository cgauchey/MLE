```python
import pandas as pd
file_url = 'https://github.com/fenago/MLEssentials/raw/main/datasets/Online%20Retail.xlsx'
df = pd.read_excel(file_url)
```

```python
df.head()
```

```python
df.sample(3).T
```

```python
# Is any thing duplicated?
df.duplicated()
```

```python
df.duplicated().sum()
```

```python
df.duplicated().value_counts()
```

```python
df[['InvoiceNo', 'StockCode', 'InvoiceDate', 'CustomerID']]
```

```python
# All the full row duplicates
df[df.duplicated()]
```

```python
df.loc[df.duplicated(), ['InvoiceNo', 'StockCode', 'InvoiceDate', 'CustomerID']]
# something will only show "true" as duplicate AFTER  the first instance
# i.e. if there are 3 duplicates, only the last 2 will be marked as duplicates
```

```python
df.loc[df.duplicated(keep='last'), ['InvoiceNo', 'StockCode',  'InvoiceDate', 'CustomerID']]
# keep last flag will NOT flag last as duplicate, but instead mark the previous ones as duplicates
# i.e. if there are 3 duplicates, only the first 2 will be marked as duplicates
```

```python
df.loc[df.duplicated(keep=False), ['InvoiceNo', 'StockCode',  'InvoiceDate', 'CustomerID']]
# will mark ALL duplicates as duplicates
```

```python
df.drop_duplicates(keep="first")
#must always run "duplicated" before running drop_duplicates
```

```python
df.duplicated().sum()
```

```python
df_no_dupes = df.drop_duplicates(keep='first')
# no in place change
# this doesn't change the original df, but instead creates a new one
# CAUTION: to edit the original df.drop_duplicates(keep='first', inplace=True)
```

```python
df_no_dupes.duplicated().sum()
```

```python
file_url = 'https://raw.githubusercontent.com/fenago/MLEssentials/main/datasets/ames_iowa_housing.csv'
df = pd.read_csv(file_url)
```

```python
df.sample(3)
```

```python
df.dtypes
```

```python
df['Id'] = df['Id'].astype('category')
```

```python
df['MSSubClass'] = df['MSSubClass'].astype('category') 
df['OverallQual'] = df['OverallQual'].astype('category') 
df['OverallCond'] = df['OverallCond'].astype('category')
```

```python
df.info()
```

```python
for col_name in ['Id','MSSubClass', 'OverallQual','OverallCond']:
    print(col_name)
    print(df[col_name].cat.categories)
```

```python
obj_df = df.select_dtypes(include='object')
# obj df that only contains variables with type object
```

```python
obj_cols = obj_df.columns
obj_cols
# only columns 
```

```python
for col_name in obj_cols:
    print(col_name)
    print(df[col_name].unique())
    
    # can find unique values in categorical values 
```

```python
for col_name in obj_cols:
    df[col_name] = df[col_name].astype('category')
```
