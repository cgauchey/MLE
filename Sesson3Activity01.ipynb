```python
import numpy as np
import pandas as pd
import altair as alt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/fenago/datasets/main/Life%20Expectancy%20Data.csv', \
                 usecols=['Status', \
                          'infant deaths', \
                          'Alcohol'])
data = data.fillna(0)
```

```python
#@title Assigning the columns to a new variable
X = data[['Alcohol', \
        'infant deaths']]
```

```python
### Perform data standardization by instantiating a StandardScaler object.
```

```python
#@title Import the MinMaxScaler and StandardScaler classes from sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
#@title Instantiate and fit MinMaxScaler with the data:
standard_scaler = StandardScaler()
standard_scaler.fit(X)
```

```python
#@title Perform the min-max scaling transformation and save the data into a new variable called X_min_max
X_standard = standard_scaler.transform(X)
X_standard
```

```python
### Analyze and define the optimal number of clusters.
```

```python
clusters = pd.DataFrame() 
clusters['cluster_range'] = range(1, 15)
inertia = []
```

```python
from sklearn.cluster import KMeans
#@title Create a for loop to go through each cluster number and fit a k-means model accordingly, 

for k in clusters['cluster_range']:
    kmeans = KMeans(n_clusters=k).fit(X_standard)
    inertia.append(kmeans.inertia_)
```

```python
#@title Assign the inertia list to a new column called 'inertia'
clusters['inertia'] = inertia
clusters
```

```python
#@title use mark_line() and encode() from the altair package to plot the Elbow graph with 'cluster_range'
alt.Chart(clusters).mark_line()\
   .encode(alt.X('cluster_range'), alt.Y('inertia'))
```

```python
#@title identify the optimal number of clusters, and assign this value to a variable called optim_cluster:
optim_cluster = 3
```

```python
#@title Train a k-means model with this number of clusters and a random_state value of 42 using the fit method from sklearn
kmeans = KMeans(random_state=42, n_clusters=optim_cluster)
kmeans.fit(X)
```

```python
#@title using the predict method from sklearn, get the predicted assigned cluster for each data point 
data['cluster2'] = kmeans.predict(X)
```

```python
alt.Chart(data).mark_circle()\
             .encode\
              (x='infant deaths', \
               y='Alcohol', \
               color='cluster2:N', \
               tooltip=['Status', 'cluster2', \
                        'infant deaths',\
'Alcohol'])\
             .interactive()
```
